A fin de tener una métrica más robusta vamos a calcular tambien la distancia de mahalanobis

```python
df_num = ds_2018[['Tem_agua','OD','pH','Turbiedad_NTU','escher_coli_ufc_100ml','Nitrato_mg_l','NH4_mg_l','P_total_l_mg_l','DBO_mg_l','Hidr_Deriv_Petr_ug_l','Cr_total_mg_l','Cd_total_mg_l','Microcistina_ug_l','Clorofila_a_ug_l','ICA']]
df_num.info()
```

```python
df_num= df_num.fillna(0)
```

```python
#ds_num.info()
df_num= df_num.astype(np.int64)
df_num.info()
```

```python
import numpy as np
import pandas as pd
import scipy as stats
from scipy.stats import chi2
  
# calculateMahalanobis Function to calculate
# the Mahalanobis distance
def calculateMahalanobis(y=None, data=None, cov=None):
  
    y_mu = y - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(y_mu, inv_covmat)
    mahal = np.dot(left, y_mu.T)
    return mahal.diagonal()
```

```python

# Creating a new column in the dataframe that holds
# the Mahalanobis distance for each row
#df['Mahalanobis'] = calculateMahalanobis(y=df, data=df[['Nitrato_mg_l','NH4_mg_l','DBO_mg_l','colif_fecales_ufc_100ml','escher_coli_ufc_100ml','enteroc_ufc_100ml','P_total_l_mg_l','Clorofila_a_ug_l','Microcistina_ug_l','ICA','Hidr_Deriv_Petr_ug_l','Cr_total_mg_l','Cd_total_mg_l','OD','pH','Tem_agua','Turbiedad_NTU']])


df_num['Mahalanobis'] = calculateMahalanobis(y=df_num, data=df_num[['Tem_agua','OD','pH','Turbiedad_NTU','escher_coli_ufc_100ml','Nitrato_mg_l','NH4_mg_l','P_total_l_mg_l','DBO_mg_l','Hidr_Deriv_Petr_ug_l','Cr_total_mg_l','Cd_total_mg_l','Microcistina_ug_l','Clorofila_a_ug_l','ICA']])  

```

```python
# calculate p-value for each mahalanobis distance considerando 3 vecinos
df_num['p'] = 1 - chi2.cdf(df_num['Mahalanobis'], 3)
  
# display first five rows of dataframe
print(df_num)
```

```python
df_num_outliers = df_num[df_num['p'] <= 0.001] # nos quedamos con los menores a 0001 considerando que los mismos son los outliers

```

The lower the Mahalanobis Distance, the closer a point is to the set of benchmark points.

A Mahalanobis Distance of 1 or lower shows that the point is right among the benchmark points. This is going to be a good one. The higher it gets from there, the further it is from where the benchmark points are.
